```python
import pandas as pd

from src.config import PATHS
from src.utils import ensure_dir
from src.recommender import train_tfidf, save_model

ensure_dir(PATHS.MODELS_DIR)

profiles = pd.read_parquet(PATHS.PROFILES_PARQUET)
corpus = pd.read_parquet(PATHS.CORPUS_PARQUET)

profiles.shape, corpus.shape
```

```python
vectorizer, tfidf_matrix, index = train_tfidf(corpus)

save_model(
    vectorizer=vectorizer,
    tfidf_matrix=tfidf_matrix,
    index=index,
    vectorizer_path=PATHS.TFIDF_VECTORIZER,
    matrix_path=PATHS.TFIDF_MATRIX,
    index_path=PATHS.RESTAURANT_INDEX,
)

print("Saved vectorizer:", PATHS.TFIDF_VECTORIZER)
print("Saved matrix:", PATHS.TFIDF_MATRIX)
print("Saved index:", PATHS.RESTAURANT_INDEX)
```

```python
from src.recommender import recommend_similar_restaurants

seed = profiles["Restaurant"].iloc[0]
recs = recommend_similar_restaurants(seed, profiles, tfidf_matrix, index, top_n=10)

seed, [r.restaurant for r in recs[:5]]
```
